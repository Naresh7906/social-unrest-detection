```python
import pandas as pd
import requests
import os
import time
```

```python
df1 = pd.read_csv('unres_id_2.csv')
df1
```

```python
bearer_token = os.environ.get("BEARER_TOKEN")
```

```python
def bearer_oauth(r):
    r.headers["Authorization"] = f"Bearer {bearer_token}"
    r.headers["User-Agent"] = "v2FilteredStreamPython"
    return r
```

```python
pos = df1[df1['tweet_id']==763298733396287488].index.values
pos
```

```python
import re
file_out = open('unrest_2.csv','a',encoding="utf8")
    
for id in df1.iloc[3092:,:].values:
        vals = []
        response = requests.get(f"https://api.twitter.com/2/tweets/{id[0]}", auth=bearer_oauth, params={'expansions':'author_id','tweet.fields':'text'})
        if response.status_code != 200:
                if response.status_code == 429:
                        print("limit reached : waiting....")
                        time.sleep(900)
                else:
                        print("Cannot get stream (HTTP {}): {}".format(response.status_code, response.text))
                        input()
        result = response.json()
        if 'data' in result:
                print(id[0],end=' ')
                text = str(re.sub(r"[^a-zA-Z0-9]", " ", result['data']['text']))
                if ',' in text:
                        text.remove(',')
                if file_out.write(f'"{text}" , {id[1]}\n'):
                        print('added')
                else:
                        print('not added')
                        break
```
